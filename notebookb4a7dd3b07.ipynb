# Importing required libraries ðŸ“š

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style("darkgrid")
```

```python
space_df = pd.read_csv(data_dir + '/Space_Corrected.csv')
```

```python
space_df
```

```python
space_df.info()
```

```python
space_df['DateTime'] = pd.to_datetime(space_df['Datum'])
space_df['Year'] = space_df['DateTime'].apply(lambda datetime: datetime.year)
```

```python
space_df["Country"] = space_df["Location"].apply(lambda location: location.split(", ")[-1])
```

```python
space_df.head()
```

```python
space_df.shape
```

```python
space_df.drop(['Unnamed: 0', 'Unnamed: 0.1'], axis = 1, inplace = True)
```

```python
space_df
```

```python
import jovian
```

```python
jovian.commit()
```

# Exploratory Analysis and Visualization

```python
#Mission Status
plt.figure(figsize=(6,6))
ax = sns.countplot(y="Status Mission", data=space_df, order=space_df["Status Mission"].value_counts().index, palette="Set2")
ax.set_xscale("log")
ax.axes.set_title("Mission Status V/S Count",fontsize=18)
ax.set_xlabel("Count",fontsize=16)
ax.tick_params(labelsize=12)
plt.tight_layout()
plt.show()
```

**From the graph** - It can be seen that Success Rate is greater than the Failure Rate

```python
count = space_df.groupby('Year')[['Status Mission']].count()
count
```

```python
plt.figure(figsize=(10,5))
plt.title("Mission Status")
sns.heatmap(count);
```

It is clearly visible that from 1965 - 1978, there were more successful missions across the world

```python
# Country vs Number of launches
plt.figure(figsize=(8,8))
ax = sns.countplot(y="Country", data=space_df, order=space_df["Country"].value_counts().index)
ax.set_xscale("log")
ax.axes.set_title("Country v/s Launches",fontsize=16)
ax.set_xlabel("Number of Launches",fontsize=15)
ax.set_ylabel("Country",fontsize=14)
ax.tick_params(labelsize=14)
plt.tight_layout()
plt.show()
```

**It can be clearly visualize that RUSSIA AND USA are doing more and more launches**

# Asking and Answering Questions

# Q1: Which Status of Rocket is more by almost every country?

```python
# Rocket status vs Count
plt.figure(figsize=(6,6))
ax = sns.countplot(x="Status Rocket", data=space_df, order=space_df["Status Rocket"].value_counts().index, palette="pastel")
ax.axes.set_title("Rocket Status V/S Count",fontsize=16)
ax.set_xlabel("Rocket Status",fontsize=15)
ax.set_ylabel("Count",fontsize=15)
ax.tick_params(labelsize=10)
plt.tight_layout()
plt.show()
```

#### Rocket Status of most of the companies is "Retired"

# Q2: Show Number of Launches per Year

```python
# Year vs Number of launches
plt.figure(figsize=(8,18))
ax = sns.countplot(y=space_df['Year'])
ax.axes.set_title("Year V/s Launches",fontsize=16)
ax.set_xlabel("Launches",fontsize=14)
ax.set_ylabel("Year",fontsize=14)
ax.tick_params(labelsize=10)
plt.tight_layout()
plt.show()
```

#### Through the Graph it can be seen that, in the Year 1971, Number of Launches is almost 118 or 119

# Q3: In which Location number of Launches is more?

```python
# Location vs Number of launches
import warnings
warnings.filterwarnings("ignore")
plt.figure(figsize=(12,56))
ax = sns.countplot(y="Location", data=space_df, order=space_df["Location"].value_counts().index)
ax.set_xscale("log")
ax.axes.set_title("Location V/S Launches",fontsize=18)
ax.set_xlabel("Number of Launches",fontsize=16)
ax.set_ylabel("Location",fontsize=16)
ax.tick_params(labelsize=12)
plt.tight_layout()
plt.show()
```

# Q4: How many Launches done by the Countries?

```python
# Company vs Number of launches
plt.figure(figsize=(8,18))
ax = sns.countplot(y="Company Name", data=space_df, order=space_df["Company Name"].value_counts().index)
ax.set_xscale("log")
ax.axes.set_title("Company V/S Launches",fontsize=18)
ax.set_xlabel("Number of Launches",fontsize=16)
ax.set_ylabel("Company Name",fontsize=16)
ax.tick_params(labelsize=12)
plt.tight_layout()
plt.show()
```
